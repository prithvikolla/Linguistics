```python
import pandas as pd
d = pd.read_csv("tweets.csv",sep='\t',names=['Cleaned'])
print(d)
```

```python
# remove special characters,numbers
d['Cleaned'] = d['Cleaned'].str.replace("[^a-zA-Z#]", " ")
d['Cleaned'].head(n=100)
```

```python
#remove punctuations
import re
import string
d['Cleaned']=d['Cleaned'].str.replace('[{}]'.format(string.punctuation), '')
d['Cleaned'].head(n=100)
```

```python
from nltk.corpus import stopwords
#remove stop words
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')

d['Cleaned']=d['Cleaned'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))
d['Cleaned'].head(n=100)
```

```python
#lower  casing
d['Cleaned']=d['Cleaned'].str.lower()
d['Cleaned'].head(n=100)
```

```python
#stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()

d['Cleaned']=d['Cleaned'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
d['Cleaned'].head(n=100)
```

```python
#lemmalization
from textblob import Word

import nltk
nltk.download('wordnet')
d['Cleaned']=d['Cleaned'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
d['Cleaned'].head(n=100)
```

```python
d.to_csv("Clean_Text_Data.csv",index= False)
```

**NAMED ENTITY RECOGNITION**

```python
j = 0
k = 0
l = 0
m = 0
n = 0
o = 0 
import spacy  
nlp = spacy.load('en_core_web_sm')
for i in range(0,98):
  text = nlp(df3['Cleaned'][i]) 
  for ent in text.ents:
    if ent.label_ == "DATE":
       j=j+1
    elif ent.label_ == "TIME":
       k=k+1
    elif ent.label_ == "CARDINAL":
       l=l+1
    elif ent.label_ == "ORDINAL":
       m=m+1
    elif ent.label_ == "NORP":
       n=n+1
    elif ent.label_ == "ORG":
       o=o+1
    else: 0
    print(ent.text, ent.label_)
print('-----------------------------------------')
print('-----------------------------------------')
print("DATE:",j) 
print("TIME:",k) 
print("CARDINAL:",l) 
print("ORDINAL:",m) 
print("NORP:",n) 
print("ORG:",o) 
```

1. ORDINAL:     ‚Äúfirst‚Äù, ‚Äúsecond‚Äù, etc.
2.   CARDINAL:    Numerals that do not fall under another type.
3.   ORG:         Companies, agencies, institutions, etc.
4.   DATE:        Absolute or relative dates or periods.
5.  TIME:        Times smaller than a day.
6.   NORP:        Nationalities or religious or political groups.
